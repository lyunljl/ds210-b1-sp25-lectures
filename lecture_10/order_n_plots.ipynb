```python
import matplotlib.pyplot as plt
import numpy as np

n = np.arange(1, 61, dtype=np.float64)
y1 = np.float64(2)**n
y2 = np.power(n, 10, dtype=np.float64)

n = np.arange(1, 61)
y1 = 2**n
print(y1.shape)
print(y1)

y2 = np.power(n, 10)
print(y2.shape)
print(y2)

plt.figure(figsize=(10, 6))
plt.plot(n, y1, label='$2^n$')
plt.plot(n, y2, label='$n^{10}$')
plt.yscale('log')
plt.xlabel('n')
plt.ylabel('Value')
plt.title('Log Plot of $2^n$ versus $n^{10}$')
plt.legend()
plt.grid(True)
plt.show()
```
